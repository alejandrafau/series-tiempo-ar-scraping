```python
from fred import Fred
```

```python
with open("fred_api_key.txt", "rb") as f:
    api_key = f.read().strip()
```

```python
fr = Fred(api_key=api_key)
```

```python
params = {
         'limit': 5,
         'output_type': 1
         }
```

```python
res = fr.series("observations", series_id="A191RL1Q225SBEA", params=params)
```

```python
res
```

```python
params = {
         'limit':2,
         'tag_names':'trade;goods'
         }

res = fr.category.series("A191RL1Q225SBEA", params=params)

for record in res:
    print(record)
```
